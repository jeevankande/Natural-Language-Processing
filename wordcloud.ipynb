```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
with open ('H://study//360Digitmg//NLP//Dataset/oneplusseven.txt','r',encoding='utf-8')as sw:
    oneplus = sw.read()
```

```python
oneplus
```

```python
other_stop_words_to_remove = ['\\n','n','\\','>','nlines','nI',"n'","hi"]
```

```python
STOPWORDS  = STOPWORDS.union(set(other_stop_words_to_remove))
```

```python
stopwords = set(STOPWORDS)
```

```python
stopwords
```

```python
text = str(oneplus)
```

```python
text
```

```python
wordcloud = WordCloud(width=2000,height=2000,
                     background_color='black',
                     max_words=200,
                     stopwords=stopwords,
                     min_font_size=12)
```

```python
wordcloud.generate(text)
```

```python
plt.imshow(wordcloud.generate(text),cmap='jet',interpolation='hamming')
plt.show()
```
